## Fit multiple polynomials

```julia
y = [y_sin,y_cos];
```

```julia
poly_list = [Polynomials.polyfit(xs, column, 4) for column in y]
```

### Normalise

```julia
using Statistics
```

```julia
y_mean = mean.(y)
```

```julia
y_std = std.(y)
```

```julia
y_norm = [(y[i].-y_mean[i])./y_std[i] for i in range(1,stop=length(y))];
```

```julia
poly_list = [Polynomials.polyfit(xs, column, 4) for column in y_norm]
```

```julia
y_norm_fit = [Polynomials.polyval(poly_elem,xs) for poly_elem in poly_list];
y_fit =[y_norm_fit[i].*y_std[i].+y_mean[i] for i in range(1,stop=length(y))];
```

```julia
scatter(xs,y,linewidth=2,title="My Plot")
plot!(xs,y_fit,linewidth=2)
```

## Functions plus simple test

```julia
import Polynomials
using Statistics
using Plots
plotly() 
```

### define functions 

```julia
"""
hej med dig
"""
function polyfit_sv(y_list, t; poly_degree=4, do_scale=1)
    y_mean = mean.(y_list)
    y_std = std.(y_list)
    
    if do_scale==1
        y_norm = [(y[i].-y_mean[i])./y_std[i] for i in range(1,stop=length(y_list))]
        poly_list = [Polynomials.polyfit(t, column,poly_degree) for column in y_norm]
        return poly_list, y_mean, y_std 
    elseif do_scale==0
        poly_list = [Polynomials.polyfit(t, column, poly_degree) for column in y_list]
        return poly_list
    else
        println("Error: do_scal can be 1 or 0")
        return
    end
end
```

```julia
function polyval_sv(poly_list,t)
    y_fit = [Polynomials.polyval(poly_elem,t) for poly_elem in poly_list]
    return y_fit
end

function polyval_sv(poly_list,t,y_mean,y_std)
    y_norm_fit = polyval_sv(poly_list,t)
    y_fit = [y_norm_fit[i].*y_std[i].+y_mean[i] for i in range(1,stop=length(y_norm_fit))];
    return y_fit
end
    
```

```julia
?polyfit_sv
```
